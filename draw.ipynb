```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

layer = [i for i in range(2, 14)]
correlation_1 = [0.2885, 0.3961, 0.2553, 0.1981, 0.1546, 0.1426, 0.0856, 0.0925, 0.0905, 0.0939, 0.0934, 0.2010]
correlation_2 = [0.2914, 0.1701, 0.2082, 0.5182, 0.6766, 0.7913, 0.8310, 0.8415, 0.8435, 0.8181, 0.7809, 0.4902]
correlation_1_bert = [0.0846, 0.0164, -0.006, 0.0051, 0.0175, -0.0042, 0.0292, 0.0052, -0.0332, -0.0485, -0.0376, -0.0729]
correlation_2_bert = [-0.2096, -0.1733, -0.3046, -0.2066, -0.1426, 0.1159, 0.4351, 0.5350, 0.5736, 0.5858, 0.5136, 0.2950]
```

```python
# Set the style
# Set the style without gridlines
sns.set(style='white', context='talk')
# Set font to sans-serif
plt.rcParams['font.family'] = 'Product Sans'

# Define custom colors
color1 = '#E9B57C'  # First color
color2 = '#E8DDC9'  # Second color
dash_color = '#B8535A'  # Color for the horizontal dashed line at y=0

# Plot lines
plt.figure(figsize=(12, 8))

# Plot MatBERT correlations
plt.plot(layer, correlation_1, label="MatBERT Correlation 1", color=color1, marker='o', linewidth=2)
plt.plot(layer, correlation_2, label="MatBERT Correlation 2", color=color2, marker='o', linewidth=2)

# Plot BERT correlations
plt.plot(layer, correlation_1_bert, label="BERT Correlation 1", linestyle="--", color=color1, marker='s', linewidth=2)
plt.plot(layer, correlation_2_bert, label="BERT Correlation 2", linestyle="--", color=color2, marker='s', linewidth=2)

# Horizontal line at y=0
plt.axhline(y=0, color=dash_color, linestyle='--', linewidth= 3)

# Labels and legend
plt.xlabel("Layer", fontsize=14, fontname='sans-serif')
plt.ylabel("Spearman's Correlation", fontsize=14, fontname='sans-serif')
plt.legend(fontsize=12)

ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Adjust layout
plt.tight_layout()

# Save and show
plt.savefig('context-average-results.png', dpi=300)
plt.show()
```
