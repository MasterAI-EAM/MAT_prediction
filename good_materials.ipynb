```python
import pandas as pd

df = pd.read_csv('Perovskite_database_content_all_data.csv')
etl = list(df['ETL_stack_sequence'])
htl = list(df['HTL_stack_sequence'])
pce = list(df['JV_default_PCE'])
```

```python
import json

with open('pair_list.json', 'r', encoding='utf-8') as f:
    pair = json.load(f)
with open('mat_dict.json', 'r', encoding='utf-8') as f:
    mat = json.load(f)
```

```python
for i, e in enumerate(htl):
    if 'SAM' in e:
        print(e, pce[i])
```

```python
def normalize(split_m, abb, mat):
    key = ['PCBB', 'SAM', 'GDY', 'PS', 'PEO', 'PCBM', 'EDOT']
    for a in abb:
        if split_m in a['full']:
            return a['max_abb']
    if '-' in split_m:
        if split_m.count('-') == 1:
            idx_ = split_m.index('-')
            return split_m[:idx_]
        else:
            ss = split_m.split('-')
            for s in ss:
                if s in key:
                    if s == 'EDOT':
                        return 'PEDOT'
            return split_m
    return split_m
```

```python
etl_dict = {}
for i, e in enumerate(htl):    
    # print(e, pce[i])
    split_e = e.split(' | ')
    for se in split_e:
        if '; ' not in se:
            norm = normalize(se, pair, mat)
            if norm not in etl_dict.keys():
                etl_dict[norm] = {}
                etl_dict[norm]['score'] = [pce[i]]
                etl_dict[norm]['origin'] = [se]
            else:
                etl_dict[norm]['score'].append(pce[i])
                etl_dict[norm]['origin'].append(se)
        else:
            for ss in se.split('; '):
                norm = normalize(ss, pair, mat)
                if norm not in etl_dict.keys():
                    etl_dict[norm] = {}
                    etl_dict[norm]['score'] = [pce[i]]
                    etl_dict[norm]['origin'] = [ss]
                else:
                    etl_dict[norm]['score'].append(pce[i])
                    etl_dict[norm]['origin'].append(ss)
sorted_etl = sorted(etl_dict.items(), key=lambda item:max(item[1]['score']), reverse=True)
for s in sorted_etl:
    print(s[0], max(etl_dict[s[0]]['score']), set(etl_dict[s[0]]['origin']))
```

```python
from mat2vec.processing import MaterialsTextProcessor

text_processor = MaterialsTextProcessor()
tmp = {}
with open('htl_score.txt', 'w', encoding='utf-8') as f:
    for s in sorted_etl:
        if s[0] in mat.keys():
            f.write(s[0]+'\t'+str(max(etl_dict[s[0]]['score']))+'\n')
        else:
            TMP = text_processor.process(s[0])[0][0]
            if TMP in mat.keys():
                f.write(TMP+'\t'+str(max(etl_dict[s[0]]['score']))+'\n')
            else:
                print(s[0], max(etl_dict[s[0]]['score']), set(etl_dict[s[0]]['origin']))
```
